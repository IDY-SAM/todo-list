```python
task_lists = []

def write_to_file(task):
    with open("todo.txt", "w") as file:
        file.write(task + "\n")

def add_task(task):
    task_lists.append({'task': task, 'completed': False})
    write_to_file(str(task_lists))
    print("Task added:", task)

def mark_completed(task_index):
    try:        
        if task_index > -1 and task_index < len(task_lists):
            task = task_lists[task_index]
            task["completed"] = True
            print(task["task"] + " marked as completed!")
        else:
            print("Task does not exist")
    except:
        print("Invalid task.")

def delete_task(task_index):
    try:
        del task_lists[task_index]
        print("Task deleted!")
    except:
        print("Invalid task.")

def show_tasks():
    print("Tasks from array:")
    for task in task_lists:
        print(task)

    file = open("todo.txt", "r")
    print("Tasks from file:")
    for task in file:
        print(task)

def add():
    while True:
        task = input("Add task or type exit to quit")
        if task == 'exit':
            break
        add_task(task)

def mark():
    while True:
        task_index = input("input task index or type exit to quit")
        if task_index == 'exit':
            break
        else:
            mark_completed(int(task_index))

def delete():
    while True:
        task_index = input("input task index or type exit to quit")
        if task_index == 'exit':
            break
        else:
            delete_task(int(task_index))

while True:
    action = input("select action: add, show, done, delete or exit to quit")

    match action:
        case 'add':
            add()
        case 'show':
            show_tasks()
        case 'done':
            mark()
        case 'delete':
            delete()
        case _:
            print('exiting')
            break
```
